```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By

# Custom user-agent
user_agent = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0"

# Set Options
options = Options()
options.set_preference("general.useragent.override", user_agent)
options.headless = True  # Run browser in headless mode (without GUI)

# Set up WebDriver with the options
driver = webdriver.Firefox(options=options)

# Initial URL of the Craigslist page
base_url = 'https://chicago.craigslist.org/search/sss?query=computers'
current_url = base_url

data = []
count = 0
max_count = 1200

while count < max_count:
    driver.get(current_url)
    print(f"Scraping page: {current_url}")
    time.sleep(5)  # adjust sleep time as needed

    listings = driver.find_elements(By.CSS_SELECTOR, 'li.cl-search-result')
    for idx, _ in enumerate(listings):
        # Re-fetch the listings to avoid StaleElementReferenceException
        listings = driver.find_elements(By.CSS_SELECTOR, 'li.cl-search-result')
        if idx >= len(listings):
            break

        link = listings[idx].find_element(By.CSS_SELECTOR, 'a.main').get_attribute('href')
        driver.get(link)
        time.sleep(3)  # wait for the page to load

        title = driver.find_element(By.CSS_SELECTOR, 'span#titletextonly').text
        description = driver.find_element(By.CSS_SELECTOR, '#postingbody').text
        image_url = driver.find_element(By.CSS_SELECTOR, 'img').get_attribute('src') if driver.find_elements(By.CSS_SELECTOR, 'img') else None

        data.append([title, link, description, image_url])
        count += 1
        print(f"Product {count} added")  # Print statement for each product added

        # Go back to the listings page
        driver.back()
        time.sleep(3)

    # Find and update the next page URL
    next_page_buttons = driver.find_elements(By.CSS_SELECTOR, 'button.bd-button.cl-next-page')
    if next_page_buttons:
        # Using JavaScript to click the button as it may not be a direct link
        driver.execute_script("arguments[0].click();", next_page_buttons[0])
        time.sleep(3)
        current_url = driver.current_url
    else:
        print("No more pages found.")
        break

# Close the WebDriver
driver.quit()

# Writing data to a CSV file
csv_file = 'craigslist_data.csv'
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Link', 'Description', 'Image URL'])
    for entry in data:
        writer.writerow(entry)

print(f"Data written to {csv_file}")
```
